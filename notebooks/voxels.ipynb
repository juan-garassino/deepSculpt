
# 3D voxel / volumetric plot

Demonstrates plotting 3D volumetric objects with `.Axes3D.voxels`.

```python
import matplotlib.pyplot as plt
import numpy as np


# prepare some coordinates
x, y, z = np.indices((8, 8, 8))

# draw cuboids in the top left and bottom right corners, and a link between
# them
cube1 = (x < 3) & (y < 3) & (z < 3)
cube2 = (x >= 5) & (y >= 5) & (z >= 5)
link = abs(x - y) + abs(y - z) + abs(z - x) <= 2

# combine the objects into a single boolean array
voxelarray = cube1 | cube2 | link

# set the colors of each object
colors = np.empty(voxelarray.shape, dtype=object)
colors[link] = 'blue'#'105,105,105'
colors[cube1] = 'blue'#105,105,105'
colors[cube2] = 'blue'#'105,105,105'

# and plot everything
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(voxelarray, facecolors=colors, edgecolor='k')


fig, axes = plt.subplots(
    ncols=2,
    nrows=2,
    figsize=(10, 10),
    #facecolor=(style),
    subplot_kw=dict(projection="3d"),
    #dpi=self.dpi,
)

axes = axes.ravel()  # flats

axes[0].voxels(
    voxelarray,
    edgecolors="k",
    #linewidth=0.1,
    facecolors=colors)
    
axes[1].voxels(
    np.rot90(voxelarray, 1),
    facecolors=np.rot90(colors, 1),
    edgecolors="k",
    #linewidth=0.08,
)

axes[2].voxels(
    np.rot90(voxelarray, 2),
    facecolors=np.rot90(colors, 2),
    edgecolors="k",
    #linewidth=0.05,
)

axes[3].voxels(
    np.rot90(voxelarray, 3),
    facecolors=np.rot90(colors, 3),
    edgecolors="k",
    #linewidth=0.02,
)

plt.show()
```
